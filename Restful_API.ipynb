Q 1 -  What is a RESTful API ?

  Ans - A RESTful API (Representational State Transfer API) is a way for different software applications to communicate over the internet using standard HTTP methods and a set of architectural principles.

Think of it like a restaurant:

Client = Customer placing an order
API = Waiter taking the order
Server = Kitchen preparing the food
Response = Food delivered back to the customer
Key Principles of a RESTful API
Uses HTTP methods
GET → Retrieve data
POST → Create new data
PUT → Update existing data
DELETE → Remove data
Resources are identified by URLs
GET /users
GET /users/123
POST /users
Stateless
Each request contains all the information needed.
The server does not remember previous requests from the client.

Data is usually exchanged in JSON

```python
{
  "id": 123,
  "name": "Alice",
  "email": "alice@example.com"
}
```

Q 2 -  What is Flask, and why is it popular for building APIs?

  Ans - Flask is a lightweight, open-source web framework for Python that is used to build web applications and APIs. It provides the basic tools needed to handle web requests, define routes, and return responses, while allowing developers to choose additional libraries as needed.

Why Flask is popular for building APIs
Simple and Lightweight
Flask has a minimal core, making it easy to learn and use.
Developers can start building APIs with very little setup.
Flexible
Flask does not enforce a specific project structure.
Developers can organize applications according to their needs.
Easy REST API Development
It provides straightforward routing and request handling.
JSON responses can be created easily, which is essential for APIs.
Extensible
Many extensions are available for authentication, database integration, API documentation, and more.
Examples include SQLAlchemy for databases and JWT libraries for authentication.
Python Ecosystem
Flask benefits from Python's extensive libraries and strong community support.
Example of a Simple Flask API

```python
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/hello', methods=['GET'])
def hello():
    return jsonify({"message": "Hello, World!"})

if __name__ == '__main__':
    app.run(debug=True)
```

Q 3 - What are HTTP methods used in RESTful APIs

  Ans - HTTP methods (also called HTTP verbs) define the action a client wants to perform on a resource in a RESTful API.
  | Method     | Purpose                                           | Example           |
| ---------- | ------------------------------------------------- | ----------------- |
| **GET**    | Retrieve data                                     | `GET /users`      |
| **POST**   | Create a new resource                             | `POST /users`     |
| **PUT**    | Update an existing resource (replace it entirely) | `PUT /users/1`    |
| **PATCH**  | Partially update a resource                       | `PATCH /users/1`  |
| **DELETE** | Remove a resource                                 | `DELETE /users/1` |

Q 4 - What is the purpose of the @app.route() decorator in Flask

  Ans - In Flask, the @app.route() decorator is used to map a URL to a Python function. When a user visits a specific URL, Flask calls the associated function and returns its result as the response.

Purpose of @app.route()
Defines a URL endpoint (route).
Connects the URL to a view function.
Specifies which HTTP methods are allowed for that route.
Example

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Welcome to the homepage!"
```

Q 5 - What is the role of Flask-SQLAlchemy

  Ans - Flask-SQLAlchemy is an extension for Flask that simplifies working with databases by integrating the SQLAlchemy ORM (Object-Relational Mapping) library into Flask applications.

Role of Flask-SQLAlchemy
Database Integration
Connects your Flask application to databases such as MySQL, PostgreSQL, SQLite, and others.
ORM Support
Allows you to work with database tables as Python classes and objects instead of writing raw SQL queries.
Simplified Database Operations
Makes it easier to create, read, update, and delete (CRUD) data.
Model Management
Lets you define database tables using Python classes called models.
Example
Configure the Database

```python
from flask import Flask
from flask_sqlalchemy import SQLAlchemy

app = Flask(__name__)
app.config["SQLALCHEMY_DATABASE_URI"] = "sqlite:///users.db"

db = SQLAlchemy(app)
```

Create a Model

```python
class User(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    name = db.Column(db.String(100), nullable=False)
```

Q 7 - How do you return JSON responses in Flask

  Ans - In Flask, you return JSON responses using the jsonify() function from the flask module. It converts Python dictionaries or lists into JSON format and automatically sets the response's content type to application/json.

Q 8 - How do you handle POST requests in Flask


  Ans - In Flask, you handle POST requests by specifying methods=['POST'] (or including POST among allowed methods) in your route and then accessing the submitted data through the request object.

Basic Example

```python
from flask import Flask, request

app = Flask(__name__)

@app.route('/submit', methods=['POST'])
def submit():
    name = request.form.get('name')
    return f"Hello, {name}!"
```

If a client sends:

```python
POST /submit
Content-Type: application/x-www-form-urlencoded

name=Alice
```

Q 9 How do you handle errors in Flask (e.g., 404)

  Ans - Flask lets you handle errors by registering error handlers with the @app.errorhandler() decorator. This allows you to return custom pages or JSON responses instead of Flask's default error pages.

Handling a 404 Error

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.errorhandler(404)
def page_not_found(error):
    return render_template('404.html'), 404
```

When a user visits a route that doesn't exist, Flask will render 404.html and return an HTTP 404 status code.

Handling Multiple Error Types

```python
@app.errorhandler(403)
def forbidden(error):
    return "Access denied", 403

@app.errorhandler(500)
def internal_error(error):
    return "Something went wrong", 500
```

Q 10 - How do you structure a Flask app using Blueprints

  Ans - Blueprints help you organize a Flask application into reusable, modular components. They're especially useful when your app grows beyond a few routes.

Basic Project Structure

A common layout looks like this:
myapp/
├── app.py
├── blueprints/
│   ├── __init__.py
│   ├── auth.py
│   └── blog.py
├── templates/
├── static/
└── requirements.txt

Create a Blueprint

In blueprints/auth.py:

```python
from flask import Blueprint

auth_bp = Blueprint(
    'auth',
    __name__,
    url_prefix='/auth'
)

@auth_bp.route('/login')
def login():
    return "Login Page"

@auth_bp.route('/register')
def register():
    return "Register Page"
```
